# Feature Selection

# Select KBest Method

```python
from sklearn.feature_selection import SelectKBest, f_classif

# Define the number of features to select
k_features = 18

# Separate the features (X) and the target variable (y)
X = data.drop(columns=['Credit_Score'])  # Drop the target column if it exists
y = data['Credit_Score']

# Initialize the SelectKBest object with the scoring function
selector = SelectKBest(score_func=f_classif, k=k_features)

# Fit the selector to your data
selector.fit(X, y)

# Transform the training data to include only the selected features
X_selected = selector.transform(X)

# Get the names of the selected features
selected_feature_names = X.columns[selector.get_support(indices=True)]

# Print the names of the selected features
print("Names of selected features:")
print(selected_feature_names)
```

```python
# Get the indices of the selected features
selected_indices = selector.get_support(indices=True)

selected_indices
```

```python
# Get the names of the selected features
selected_column_names = X.columns[selected_indices]

# Create a DataFrame with the selected features and the target variable
selected_features_df = pd.DataFrame(X_selected, columns=selected_column_names)
merged_data = pd.concat([selected_features_df, y], axis=1)

# Display the selected features and the target column
merged_data.head()
```

# Spliting the data into training and test data

```python
# Separate features and target column
X = merged_data.drop(columns=['Credit_Score'])  
y = merged_data['Credit_Score']
```

# Scaling the dataset

# MinMax Scaler

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit the scaler only on the training data
scaler.fit(X_train)

# Transform both training and test data using the scaler fitted on training data
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# Spliting the data into training and test data

# Pipeline the training dataset

The Pipeline where two resampling techniques are applied sequentially:

Oversampling (SMOTE): Synthetic Minority Over-sampling Technique generates synthetic samples for the minority class. sampling_strategy='auto' adjusts the number of synthetic samples to balance the classes. Undersampling (RandomUnderSampler): Randomly removes samples from the majority class to balance the classes. sampling_strategy='auto' adjusts the number of samples to be removed.

In summary, this pipeline combines oversampling (SMOTE) and undersampling (RandomUnderSampler) techniques to address class imbalance in the training data. The goal is to create a more balanced dataset for training machine learning models.

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline

# Create the resampling pipeline
pipeline = Pipeline([('oversample', SMOTE(sampling_strategy='auto')),('undersample', RandomUnderSampler(sampling_strategy='auto'))])

# Fit and resample the data
X_train_resampled, y_train_resampled = pipeline.fit_resample(X_train_scaled, y_train)
```

# Fine tuning using GridSearch technique for Random Forest Classifier

```python
from sklearn.model_selection import RandomizedSearchCV
#from scipy.stats import randint

# Create a Random Forest Classifier
rfc_new = RandomForestClassifier(random_state=0)

# Define the hyperparameter distribution
#param_dist = {
#    'n_estimators': randint(10, 200),
#    'max_depth': [None, 10, 20,30],
#    'min_samples_split': randint(2, 10),
#    'min_samples_leaf': randint(1, 4)
#}

param_dist = {
    'n_estimators': [10,50,100, 200, 300],
    'max_depth': [None, 10, 20,30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
    
# Perform Randomized Search
randomized_search = RandomizedSearchCV(rfc_new, param_distributions=param_dist, n_iter=10, cv=15, scoring='accuracy',
                                       random_state=0)

randomized_search.fit(X_train_resampled, y_train_resampled)

# Get the best hyperparameters
best_params = randomized_search.best_params_

# Train a new model with the best hyperparameters
best_rf_model = RandomForestClassifier(**best_params)
best_rf_model.fit(X_train_resampled, y_train_resampled)

# Evaluate the model on the test set
accuracy = best_rf_model.score(X_test_scaled, y_test)*100
print("Accuracy:", str(round(accuracy,2))+"%")
```

# Cross Validation (Stratified K-Fold)

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, classification_report

# Initialize StratifiedKFold with 15 folds
skf = StratifiedKFold(n_splits=15)

rfc_val = RandomForestClassifier(random_state=0)

# Lists to store cross-validation scores
cv_scores = []

# Iterate over each fold
for train_index, test_index in skf.split(X_train_resampled, y_train_resampled):
    # Split the data into train and test sets for this fold
    X_train_fold, X_test_fold = X_train_resampled[train_index], X_train_resampled[test_index]
    y_train_fold, y_test_fold = y_train_resampled[train_index], y_train_resampled[test_index]
    
    # Train your model on the training data for this fold
    rfc_val.fit(X_train_fold, y_train_fold)
    
    # Make predictions on the test data for this fold
    y_pred_fold = rfc_val.predict(X_test_fold)
    
    # Calculate accuracy score for this fold and append to the list
    fold_score = accuracy_score(y_test_fold, y_pred_fold)
    cv_scores.append(fold_score)

    # Optionally, print the classification report for this fold
    print("Classification Report for Fold", len(cv_scores))
    print(classification_report(y_test_fold, y_pred_fold))
    print("\n")

# Calculate mean accuracy across all folds
mean_accuracy = np.mean(cv_scores)
print("Mean Accuracy:", mean_accuracy)

# Print cross-validation scores for each fold
print("Cross-Validation Scores:", cv_scores)
```

```python
import os

def split_pickle_file(input_file, output_prefix, max_part_size):
    part_num = 0
    with open(input_file, 'rb') as f:
        while True:
            chunk = f.read(max_part_size)
            if not chunk:
                break
            output_file = f"{output_prefix}_{part_num}.pkl"
            with open(output_file, 'wb') as part:
                part.write(chunk)
            part_num += 1

input_file = 'random_forest_model.pkl'
output_prefix = 'model_part'
max_part_size = 60 * 1024 * 1024  # Split into approximately 60 MB parts
split_pickle_file(input_file, output_prefix, max_part_size)
```
