# Outlier detection

# Box Plot

```python
fig, ax = plt.subplots(ncols=6, nrows=4, figsize=(20, 10))
count = 0
ax = ax.flatten()

# Loop through each column in the DataFrame
for i, col in enumerate(data.columns):
    # Check if the column is numeric
    if data[col].dtype in ['int64', 'float64']:
        sns.boxplot(y=data[col], ax=ax[count])
        ax[count].set_title(f'{col}')
        #ax[count].yaxis.set_label_coords(0, 0)  # Adjust the label position
        count += 1
# Remove empty subplots if any
for i in range(count, len(ax)):
    fig.delaxes(ax[i])

plt.tight_layout(pad=0.8, w_pad=1.0, h_pad=5.0)
plt.suptitle("Box Plots", y=1.03)  # Adjust title position
plt.show()
```

```python
data['Annual_Income'].mean()
```

# Replacing the outliers with the mean of respective column

```python
# Columns with outliers
columns_with_outliers = ['Annual_Income', 'Monthly_Inhand_Salary', 'Delay_from_due_date','Changed_Credit_Limit','Num_Credit_Inquiries','Outstanding_Debt','Credit_Utilization_Ratio','Total_EMI_per_month','Amount_invested_monthly','Monthly_Balance']

# Dictionary to store boolean outliers for each column
outliers_dict = {}


# Loop through each column
for column_name in columns_with_outliers:
    # Calculate the interquartile range (IQR)
    q1 = data[column_name].quantile(0.25)
    q3 = data[column_name].quantile(0.75)
    iqr = q3 - q1

    # Define the upper and lower bounds for outliers
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    # Identify outliers in boolean format
    outliers = (data[column_name] < lower_bound) | (data[column_name] > upper_bound)

    # Store boolean outliers in the dictionary
    outliers_dict[column_name] = outliers

    # Optionally, you can replace outliers with the mean of the column
    data[column_name] = data[column_name].where(~outliers, data[column_name].mean())

# Display the new dataset
data.head()
```

# Encoding

```python
from sklearn.preprocessing import LabelEncoder

# List of categorical columns
categorical_columns = ['Type_of_Loan', 'Credit_Mix', 'Payment_of_Min_Amount', 'Payment_Behaviour','Credit_Score']

# Create a LabelEncoder instance
label_encoder = LabelEncoder()

# Apply label encoding to each categorical column
for column in categorical_columns:
    if column in data.columns:
        data[column] = label_encoder.fit_transform(data[column])

# Display the DataFrame after label encoding
data.head()
```

```python
## Deleting few columns as it won't impact the dataset.

data.drop('ID', axis=1, inplace=True)
data.drop('Customer_ID', axis=1, inplace=True)
data.drop('Month', axis=1, inplace=True)
data.drop('Name', axis=1, inplace=True)
data.drop('Age', axis=1, inplace=True)
data.drop('SSN', axis=1, inplace=True)
data.drop('Occupation', axis=1, inplace=True)

data.head()
```

# Scaling

```python
from sklearn.preprocessing import Normalizer

# Get the column name for the future use
feature_columns = data.iloc[:,:-1]
target_column=data['Credit_Score']

# Initiatiate and fit scaler
scaler = Normalizer()
data_scaled = scaler.fit_transform(feature_columns)

# Embedding the scaled data into the dataset
data_scaled_new = pd.DataFrame(data_scaled,columns=feature_columns.columns)

data_scaled_new.head()
```
